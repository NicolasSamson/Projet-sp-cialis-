## Problème 1D avec capteur pas dans les mêmes unitées




Le véhicule se dépalce donc le long de l'axe x, il comporte une mesure au laser qui calcule la position du véhicule.

Le vecteur d'état est le suivant: 
$$ \vec{x}(k) = \begin{bmatrix}
x\\
\dot{x}
\end{bmatrix} $$

Voici la liste des hypothèses: 
- Le bruit que subit l'estimation du modèle. $$C_v = \begin{bmatrix} N \sim (0,\sigma_{positionPas}^2) \\ N \sim (0,\sigma_{vitessePas}^2)  \end{bmatrix} = C_v = \begin{bmatrix} N \sim (0,1) \\ N \sim (0,1)  \end{bmatrix} $$
- Le modèle de la propagation d'état est directement proportionnel au déplacement du véhicule $$\phi= \begin{bmatrix}
1& \Delta_t\\
0& 0
\end{bmatrix} $$
- La commande est appliqué telquel au véhicule. $$\Gamma = \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix}$$ 
- La mesure du capteur z(x) est non biaisé mais subit un bruit gaussien. La mesure du capteur est donné en volt par la capteur selon les équations suivantes: 
$$C_w = \begin{bmatrix}
N \sim (0,\sigma_{laser}^2) \\
N \sim (0,\sigma_{encodeur}^2)
\end{bmatrix} =
\begin{bmatrix}
N \sim (0,4) \\
N \sim (0,4)
\end{bmatrix}
$$

$$\Lambda = \begin{bmatrix}
1 & 0\\
0 & 1
\end{bmatrix}  $$

La commande utilisée est 5 mètres par seconde $$ u = \begin{bmatrix}
0 \\
5
\end{bmatrix}  $$


```python
### Variable model 1 D 2 états (positions,vitesses)
delta_t = 0.1 
sigma_laser = 4
sigma_encodeur = 4

sigma_process_noise = 1

##
#delta_t = 0.1 
#sigma_laser = 0.5
#sigma_encodeur = 0.5

#sigma_process_noise = 1
##
x0 = np.array([[0.],    # position
    [0.]]) # velocity


f = KalmanFilter(dim_x=2, dim_z=2,dim_u =2)
f.x = np.array(x0)  

#state transition matrix Phi
phi = np.array([[1.,delta_t],
                [0.,0.]])
f.F = phi
# measurement function: alpha
alpha = np.array([[1.,0.],
                [0.,1.]])
f.H = alpha

# the covariance matrix
f.P = np.array([[1000.,    0.],
                [   0., 1000.] ])
#  measurement noise cW
f.R = np.array([[sigma_laser,    0.],
                [   0., sigma_encodeur] ])
#  Process noise cW
process_noise = Q_discrete_white_noise(dim=2, dt=delta_t, var=sigma_process_noise)
f.Q = process_noise

# command tansition Matrix
gamma = np.array([[1,    0.],
                [   0., 1] ])
f.B = gamma

u = np.array([[0.],    # position
            [5.]])   # velocity



### Variable to store 
size = 400
x0_array = np.zeros(size)
x1_array = np.zeros(size)
z0_array = np.zeros(size)
z1_array = np.zeros(size)

x_0_reel = np.zeros(size)
x_1_reel = np.zeros(size)

p_1_reel = np.zeros(size)
p_2_reel = np.zeros(size)

x = x0
x_real= x0
time_axe = np.zeros(size)

for i in range(size):


    x_real = np.dot(phi,x_real)+ np.dot(gamma,u) #+ np.reshape(np.random.normal(0, [sigma_process_noise,sigma_process_noise]),x.shape)
    
    z = np.dot(alpha,x_real) + np.reshape(np.random.normal(0, [np.sqrt(sigma_laser),np.sqrt(sigma_encodeur)]),x.shape)
    f.predict(u)
    f.update(z)

    var_estimated = f.P

    time = i*delta_t
    # Stock les variables
    x = f.x
    x0_array[i] = f.x[0]
    x1_array[i] = f.x[1]
    z0_array[i] = f.z[0]
    z1_array[i] = f.z[1]

    x_0_reel[i] = x_real[0]
    x_1_reel[i] = x_real[1]
    time_axe[i] = time

    p_2_reel[i] = var_estimated[1,1]
    p_1_reel[i] = var_estimated[0,0]

# Affichage graphique
plt.figure(figsize=(15,15))
plt.plot(time_axe,x_0_reel,label="x_reel" )
plt.scatter(time_axe,z0_array,label="x_mesure" )
plt.scatter(time_axe,x0_array,label="xKalman")
plt.errorbar(time_axe,z0_array,yerr=np.sqrt(np.ones(size)*sigma_laser),fmt="o",c="green")
plt.errorbar(time_axe,x0_array,yerr=np.sqrt(p_1_reel),fmt="o",c="orange")

plt.title("Position estimé en  fonction du temps")
plt.ylabel("Position [m]")
plt.xlabel("temps [s]")
plt.legend() 
plt.show()

plt.figure(figsize=(15,15))
plt.plot(time_axe,x_0_reel-x_0_reel,label="x_reel" )
plt.scatter(time_axe,z0_array-x_0_reel,label="x_mesure")
plt.scatter(time_axe,x0_array-x_0_reel,label="xKalman")

plt.errorbar(time_axe,z0_array-x_0_reel,yerr=np.sqrt(np.ones(size)*sigma_laser),fmt="o",c="green")
plt.errorbar(time_axe,x0_array-x_0_reel,yerr=np.sqrt(p_1_reel),fmt="o",c="orange")

plt.title("Position estimé en  fonction du temps")
plt.ylabel("Erreur [m]")
plt.xlabel("temps [s]")
plt.legend() 
plt.show()


plt.figure(figsize=(10,10))
plt.plot(time_axe,x_1_reel,label="v_reel" )
plt.scatter(time_axe,z1_array,label="v_mesure" )
plt.scatter(time_axe,x1_array,label="vKalman")
plt.title("vitesse estimé en  fonction du temps")
plt.ylabel("Vitesse [m/s]")
plt.xlabel("temps [s]")
plt.legend() 
plt.show()
```
